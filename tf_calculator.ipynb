```python
from sympy import diff, Function, laplace_transform, laplace_correspondence, laplace_initial_conds, symbols, abc, Eq, solve, simplify
```

```python
v_c, V_C, v, V = symbols('v_c, V_C, v, V', cls=Function)
s, L, R, C = symbols('s, L, R, C')
t = symbols('t', real=True)
```

```python
LHS = L*C*diff(v_c(t), t, 2) + R*C*diff(v_c(t), t) + v_c(t)
RHS = v(t)
Eq(LHS, RHS)
```

```python
lhs_lt = laplace_transform(LHS, t, s, noconds=True)
rhs_lt = laplace_transform(RHS, t, s, noconds=True)

e1 = laplace_correspondence(lhs_lt, {v_c: V_C})
e2 = laplace_correspondence(rhs_lt, {v: V})
```

```python
e3 = laplace_initial_conds(e1, t, {v_c: [0, 0, 0]})
e4 = laplace_initial_conds(e2, t, {v: [0, 0, 0]})
e5 = Eq(e3, e4)
e5
```
